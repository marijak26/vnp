```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('/home/marija26/ElectricCarData.csv')
```

```python
data.head()
```

```python
encoder = LabelEncoder()
```

```python
for column in data.columns:
    if type(data[column].to_list()[0]) == str:
         data[column] = encoder.fit_transform(data[column])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['PriceEuro'], axis=1)
Y = data['PriceEuro']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data
```

```python
model = DecisionTreeRegressor(max_depth=3,min_samples_leaf=5,max_leaf_nodes=10,min_samples_split=5)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse_dt=mean_squared_error(y_test, y_pred)
```

```python
r2_dt=r2_score(y_test, y_pred)
```

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor(n_estimators=100, max_depth=6, learning_rate=0.1, random_state=0, objective="reg:squarederror")
```

```python
xgb.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb.predict(X_test)
```

```python
r2_xgb = r2_score(y_test, y_pred_xgb)
```
